```r
dsn_driver <- "BLUDB" 
dsn_database <- "BLUDB" 
dsn_hostname <- "bluemix05.bluforcloud.com" 
dsn_port <- "50000" 
dsn_protocol <- "TCPIP" 
dsn_uid <- "dash018339" 
dsn_pwd <- "XXXXX"
```

```r
#You need to specify the driver location below
```

```r
jcc = JDBC("com.ibm.db2.jcc.DB2Driver","/usr/local/src/dataconnector-db2/db2jcc4-10.5.0.6.jar");
jdbc_path = paste("jdbc:db2://",  dsn_hostname, ":", dsn_port, "/", dsn_database, sep="");
conn = dbConnect(jcc, jdbc_path, user=dsn_uid, password=dsn_pwd)
```

```r
query = "SELECT * FROM OPD_HASHTAGS";
rs = dbSendQuery(conn, query);
df = fetch(rs, -1);
df
```
